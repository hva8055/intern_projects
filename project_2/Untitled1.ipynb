```python
from PyQt5.uic import loadUi
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import QApplication
from PyQt5.QtWidgets import QDialog
import sys

x = []

class page1(QDialog):
    
    def __init__(self):
        super(page1,self).__init__()
        loadUi('1.ui',self)
        self.next1.clicked.connect(self.p2)
        
    def p2(self):
        tenure = self.tenure.text()
        x.append(int(tenure))
        widget.setCurrentIndex(widget.currentIndex()+1)
        self.tenure.clear()
        
class page2(QDialog):
    
    def __init__(self):
        super(page2,self).__init__()
        loadUi('2.ui',self)
        self.next2.clicked.connect(self.p3)
        self.back2.clicked.connect(self.b1)
        
    def p3(self):
        ph = self.PhoneService.text()
        x.append(int(ph))
        widget.setCurrentIndex(widget.currentIndex()+1)
        self.PhoneService.clear()
        
    def b1(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)
        
class page3(QDialog):
    
    def __init__(self):
        super(page3,self).__init__()
        loadUi('3.ui',self)
        self.next3.clicked.connect(self.p4)
        self.back3.clicked.connect(self.b2)
        
    def p4(self):
        ph = self.Contract.text()
        x.append(int(ph))
        widget.setCurrentIndex(widget.currentIndex()+1)
        
    def b2(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)
        
class page4(QDialog):
    
    def __init__(self):
        super(page4,self).__init__()
        loadUi('4.ui',self)
        self.next4.clicked.connect(self.p5)
        self.back4.clicked.connect(self.b3)
        
    def p5(self):
        ph = self.PaperlessBilling.text()
        x.append(int(ph))
        widget.setCurrentIndex(widget.currentIndex()+1)
        self.PaperlessBilling.clear()
        
    def b3(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)
        
class page5(QDialog):
    
    def __init__(self):
        super(page5,self).__init__()
        loadUi('5.ui',self)
        self.next5.clicked.connect(self.p6)
        self.back5.clicked.connect(self.b4)
        
    def p6(self):
        ph = self.PaymentMethod.text()
        x.append(int(ph))
        widget.setCurrentIndex(widget.currentIndex()+1)
        self.PaymentMethod.clear()
        
    def b4(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)
        
class page6(QDialog):
    
    def __init__(self):
        super(page6,self).__init__()
        loadUi('6.ui',self)
        self.next6.clicked.connect(self.p7)
        self.back6.clicked.connect(self.b5)
        
    def p7(self):
        ph = self.MonthlyCharges.text()
        x.append(float(ph))
        widget.setCurrentIndex(widget.currentIndex()+1)
        self.MonthlyCharges.clear()
        
    def b5(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)
        
class page7(QDialog):
    
    def __init__(self):
        super(page7,self).__init__()
        loadUi('7.ui',self)
        self.predict.clicked.connect(self.pred)
        self.back7.clicked.connect(self.b6)
        self.again.clicked.connect(self.new1)
        
    def pred(self):
        global x
        import pandas as pd
        from sklearn.model_selection import train_test_split
        from sklearn.linear_model import LogisticRegression
        
        ph = self.TotalCharges.text()
        x.append(float(ph))
        data = pd.read_csv('edited.csv')
        t = data.drop(['Churn','Unnamed: 0'],axis=1)
        y = data.drop(['tenure', 'PhoneService', 'Contract', 'PaperlessBilling','PaymentMethod', 
                       'MonthlyCharges', 'TotalCharges','Unnamed: 0'], axis=1)
        x_train,x_test,y_train,y_test = train_test_split(t,y)
        model = LogisticRegression()
        model.fit(x_train,y_train)
        y_pred = model.predict([x])
        if y_pred[0]=='No':
            r = 'high risk customer'
            self.result.setText(r)
        else:
            r = 'low risk customer'
            self.result.setText(r)
                
    def b6(self):
        x.pop()
        widget.setCurrentIndex(widget.currentIndex()-1)

    def new1(self):
        self.TotalCharges.clear()
        x.clear()
        widget.setCurrentIndex(widget.currentIndex()-6)
        
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    widget = QtWidgets.QStackedWidget()
    page1 = page1()
    page2 = page2()
    page3 = page3()
    page4 = page4()
    page5 = page5()
    page6 = page6()
    page7 = page7()
    widget.addWidget(page1)
    widget.addWidget(page2)
    widget.addWidget(page3)
    widget.addWidget(page4)
    widget.addWidget(page5)
    widget.addWidget(page6)
    widget.addWidget(page7)
    widget.setFixedHeight(551)
    widget.setFixedWidth(601)
    widget.show()
    sys.exit(app.exec_())
```
